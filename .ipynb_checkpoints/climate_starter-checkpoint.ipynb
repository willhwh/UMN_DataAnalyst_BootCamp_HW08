# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
# reflect the tables
Base=automap_base()
Base.prepare(engine,reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement=Base.classes.measurement
Station=Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session=Session(engine)
```

```python
#get columns info.
from sqlalchemy import inspect
inspector=inspect(engine)

for i in Base.classes.keys():
    columns=inspector.get_columns(i)
    print(i+' TABLE has following columns\n==========')
    for column in columns:
        print(column['name'],column['type'])
    print('==========\n')
```

```python
#take a closer look at measurement TABLE data
n=0
for row in session.query(Measurement.id,Measurement.station,Measurement.date,Measurement.prcp,Measurement.tobs):
    n=n+1
    print(row)
    if n==5:
        break
```

```python
#take a closer look at station TABLE data
engine.execute('select * from station limit 5').fetchall()
```

# Exploratory Climate Analysis

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results

# Calculate the date 1 year ago from the last data point in the database
from datetime import datetime
last_date=session.query(Measurement.date).all()[-1][0]
last_date_origin=dt.datetime.strptime(last_date,"%Y-%m-%d")
last_date_p1=last_date_origin-dt.timedelta(days=365)
last_date_p1=last_date_p1.strftime('%Y-%m-%d')

# Perform a query to retrieve the data and precipitation scores
last_y=session.query(Measurement.date,Measurement.prcp)\
    .filter(Measurement.date>=last_date_p1).all()


# Save the query results as a Pandas DataFrame and set the index to the date column
last_y_df=pd.DataFrame(last_y)
last_y_df=last_y_df.set_index('date')

# Sort the dataframe by date
last_y_df.sort_values('date',inplace=True)

# Use Pandas Plotting with Matplotlib to plot the data
import matplotlib.ticker as ticker
fig, ax = plt.subplots(1,1,figsize=(10,10))

#change the xticks label frequence to show once every 30 days data
ax.plot(last_y_df,label='Prcipitation')
tick_spacing = 30


#x label rotaion
ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
plt.xticks(rotation=90)

plt.xlabel('Date')
plt.ylabel('Precipitation')
plt.legend('Precipitation')
plt.legend(prop={"size":15})
```

```python
# Use Pandas to calcualte the summary statistics for the precipitation data
last_y_df.describe()
```

```python
# Design a query to show how many stations are available in this dataset?
stations=session.query(Measurement.station,func.count(Measurement.station))\
    .group_by(Measurement.station).order_by(func.count(Measurement.station).desc()).all()

print('Total amount of Stations:\n'+ str(len(stations)))

print('\nStations:')
station_id=[station[0] for station in stations]
station_count=[station[1] for station in stations]

target=stations[0][0]
print(f'{target} is the most active station')

pd.DataFrame({"Station_ID":station_id,'Station_Count':station_count})

```

```python
# List the stations and the counts in descending order.
print(f'Temerature Info. for Station ID: {target}\n')
print('Highest Temperature')
print(session.query(func.max(Measurement.tobs)).filter(Measurement.station==target).first()[0])
print('Lowest Temperature')
print(session.query(func.min(Measurement.tobs)).filter(Measurement.station==target).first()[0])
print('Average Temperature')
print(round(session.query(func.avg(Measurement.tobs)).filter(Measurement.station==target).first()[0],1))
```

```python
# Choose the station with the highest number of temperature observations.
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
last_y_temperature=session.query(Measurement.date,Measurement.tobs)\
    .filter(Measurement.date>=last_date_p1).filter(Measurement.station==target).all()

last_y_temperature_df=pd.DataFrame(last_y_temperature)
last_y_temperature_df.set_index('date',inplace=True)

print(last_y_temperature_df.head())

plt.hist(last_y_temperature_df['tobs'],bins=12,label='tobs')
#last_y_temperature_df.plot.hist(bins=12)

plt.legend()
plt.xlabel('Temperature (F)')
```

```python
#Use the t-test to determine whether the difference in the means, 
#if any, is statistically significant. Will you use a paired t-test, or an unpaired t-test? Why?
```

```python
june_days=session.query(Measurement.tobs).filter(func.strftime("%m", Measurement.date) == '06').all()
dec_days=session.query(Measurement.tobs).filter(func.strftime("%m", Measurement.date) == '12').all()
june_total=[day[0] for day in june_days]
dec_total=[day[0] for day in dec_days]
```

```python
fig,[ax1,ax2]=plt.subplots(1,2)
ax1.hist(june_total)
ax1.title.set_text('June Temperature Distribution')
ax1.set_xlabel('Temperature')
ax1.set_ylabel('Counts')

ax2.hist(dec_total)
ax2.title.set_text('Dec Temperature Distribution')
ax2.set_xlabel('Temperature')
ax2.set_ylabel('Counts')
```

```python
june=session.query(func.avg(Measurement.tobs)).filter(func.strftime("%m", Measurement.date) == '06').all()
dec=session.query(func.avg(Measurement.tobs)).filter(func.strftime("%m", Measurement.date) == '12').all()
june_len=len(session.query(Measurement.tobs).filter(func.strftime("%m", Measurement.date) == '06').all())
dec_len=len(session.query(Measurement.tobs).filter(func.strftime("%m", Measurement.date) == '12').all())

print(f'There are {june_len} data for June Temperature and the average temperature is {june[0][0]} ')
print(f'There are {dec_len} data for December Temperature and the average temperature is {dec[0][0]} ')
```

Since the data points for June and Dec. don't match each other, so I will use unpaired t-test.

* Null Hypothesis: µa = µb (the means of both temperature are equal)
* Alternate Hypothesis: µa ≠ µb (the means of both temperature are not equal)

```python
import scipy.stats as stats
stats.ttest_ind(np.array(june_total),np.array(dec_total))
```

The pvalue is siginificantly smaller than 0.05 so we have strong evidence to rejcet the Null Hypothesis. Thus, the temeprature in June and Dec. in Hawii is different; Dec. is a bit colder the June.

# Temperature Analysis II

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()

# function usage example
print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
# Use your previous function `calc_temps` to calculate the tmin, tavg, and tmax 
# for your trip using the previous year's data for those same dates.
print('My trip start date was 2015-07-10\n')

print(calc_temps('2014-07-10', '2015-07-10'))
```

```python
# Plot the results from your previous query as a bar chart. 
# Use "Trip Avg Temp" as your Title
# Use the average temperature for the y value
# Use the peak-to-peak (tmax-tmin) value as the y error bar (yerr)
error=calc_temps('2014-07-10', '2015-07-10')[0][2]-calc_temps('2014-07-10', '2015-07-10')[0][0]
avg=calc_temps('2014-07-10', '2015-07-10')[0][1]
plt.figure(figsize=(3,5))
plt.bar(x=' ',height=avg,yerr=error,color='pink',alpha=0.5)
plt.ylabel('Temp(F)')
plt.title('Trip Avg Temp')
```

# Daily Rainfall Average

```python
# Calculate the total amount of rainfall per weather station for your trip dates using the previous year's matching dates.
# Sort this in descending order by precipitation amount and list the station, name, latitude, longitude, and elevation
start='2014-07-10'
end='2015-07-10'

sel = [Measurement.station,Station.name,Station.latitude,Station.longitude,Station.elevation]

travel_df=session.query(*sel).filter(Measurement.date>=start).filter(Measurement.station == Station.station).filter(Measurement.date<=end)\
    .group_by(Measurement.station).order_by(func.sum(Measurement.tobs).desc()).all()

pd.DataFrame(travel_df)
```

```python
# Create a query that will calculate the daily normals 
# (i.e. the averages for tmin, tmax, and tavg for all historic data matching a specific month and day)

def daily_normals(date):
    """Daily Normals.
    
    Args:
        date (str): A date string in the format '%m-%d'
        
    Returns:
        A list of tuples containing the daily normals, tmin, tavg, and tmax
    
    """
    
    sel = [func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)]
    return session.query(*sel).filter(func.strftime("%m-%d", Measurement.date) == date).all()
    
daily_normals("01-01")

```

```python
# calculate the daily normals for your trip
# push each tuple of calculations into a list called `normals`

# Set the start and end date of the trip
print('My vacation range is 2015-07-10 to 2015-07-15\n')
start = dt.datetime.strptime("2015-07-10", "%Y-%m-%d")
end = dt.datetime.strptime("2015-07-15", "%Y-%m-%d")


# Use the start and end date to create a range of dates
date_generated = [start + dt.timedelta(days=x) for x in range(0, (end-start).days)]
date_generated

# Stip off the year and save a list of %m-%d strings
date_m_d=[date.strftime("%m-%d") for date in date_generated]

# Loop through the list of %m-%d strings and calculate the normals for each date
normals=[daily_normals(date) for date in date_m_d]
normals
```

```python
min_temp=[]
avg_temp=[]
max_temp=[]
for i in normals:
    min_temp.append(i[0][0])
    avg_temp.append(i[0][1])
    max_temp.append(i[0][2])
```

```python
# Load the previous query results into a Pandas DataFrame and add the `trip_dates` range as the `date` index
date_info_df=pd.DataFrame({'Min':min_temp,'Avg':avg_temp,'Max':max_temp})
date_info_df['trip_dates']=date_m_d
date_info_df.set_index('trip_dates',inplace=True)
date_info_df                    
                           
```

```python
# Plot the daily normals as an area plot with `stacked=False`
date_info_df.plot.area(stacked=False,alpha=0.4)
plt.ylabel('Temperature(F)')
```
